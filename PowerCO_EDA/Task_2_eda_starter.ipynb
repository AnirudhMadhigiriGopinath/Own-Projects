## Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations.

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:

        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

Thhe first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

```python
import numpy as np

# Check the data types of each column
print("Data Types of Client Data:")
print(client_df.dtypes)

# Identify non-numeric columns
non_numeric_columns = client_df.select_dtypes(exclude=[np.number]).columns.tolist()
print("\nNon-Numeric Columns:")
print(non_numeric_columns)

# Drop non-numeric columns
client_df_numeric = client_df.drop(columns=non_numeric_columns)

# Convert categorical columns to numeric (if applicable)
# For example, using one-hot encoding for categorical variables
client_df_encoded = pd.get_dummies(client_df, drop_first=True)

# Calculate the correlation matrix for numeric features
correlation_matrix = client_df_numeric.corr()  # Use client_df_encoded if you chose option 2

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Client Data')
plt.show()
```

```python
# Display the first few rows of each dataset
print("Client Data:")
print(client_df.head())
print("\nPrice Data:")
print(price_df.head())

# Data Types of Each Column
print("\nData Types of Client Data:")
print(client_df.dtypes)

print("\nData Types of Price Data:")
print(price_df.dtypes)

# Descriptive Statistics of the Datasets
print("\nDescriptive Statistics of Client Data:")
print(client_df.describe(include='all'))

print("\nDescriptive Statistics of Price Data:")
print(price_df.describe(include='all'))

# Check for Missing Values
print("\nMissing Values in Client Data:")
print(client_df.isnull().sum())

print("\nMissing Values in Price Data:")
print(price_df.isnull().sum())

# Visualizing Distributions of Relevant Columns
# For Client Data
plt.figure(figsize=(12, 6))
sns.histplot(client_df['cons_12m'], bins=30, kde=True)
plt.title('Distribution of Electricity Consumption (Last 12 Months)')
plt.xlabel('Electricity Consumption (kWh)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(client_df['forecast_cons_12m'], bins=30, kde=True)
plt.title('Distribution of Forecasted Electricity Consumption (Next 12 Months)')
plt.xlabel('Forecasted Electricity Consumption (kWh)')
plt.ylabel('Frequency')
plt.show()

# For Price Data
plt.figure(figsize=(12, 6))
sns.histplot(price_df['price_off_peak_var'], bins=30, kde=True)
plt.title('Distribution of Off-Peak Variable Price')
plt.xlabel('Off-Peak Variable Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(price_df['price_peak_var'], bins=30, kde=True)
plt.title('Distribution of Peak Variable Price')
plt.xlabel('Peak Variable Price')
plt.ylabel('Frequency')
plt.show()

# Churn Analysis
plt.figure(figsize=(8, 5))
sns.countplot(x='churn', data=client_df)
plt.title('Churn Count')
plt.xlabel('Churn (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()

# Analyzing Churn by Customer Characteristics
# Boxplot of electricity consumption by churn status
plt.figure(figsize=(12, 6))
sns.boxplot(x='churn', y='cons_12m', data=client_df)
plt.title('Electricity Consumption (Last 12 Months) by Churn Status')
plt.xlabel('Churn (1 = Yes, 0 = No)')
plt.ylabel('Electricity Consumption (kWh)')
plt.show()

# Analyzing Churn by Sales Channel
plt.figure(figsize=(12, 6))
sns.countplot(x='channel_sales', hue='churn', data=client_df)
plt.title('Churn Count by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Count')
plt.legend(title='Churn', loc='upper right', labels=['No', 'Yes'])
plt.xticks(rotation=45)
plt.show()

# Price Trends Over Time
# Convert price_date to datetime
price_df['price_date'] = pd.to_datetime(price_df['price_date'])

# Plotting price trends
plt.figure(figsize=(14, 7))
sns.lineplot(data=price_df, x='price_date', y='price_off_peak_var', label='Off-Peak Variable Price')
sns.lineplot(data=price_df, x='price_date', y='price_peak_var', label='Peak Variable Price')
plt.title('Price Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Summary of Findings
print("\nSummary of Findings:")
print("1. The distribution of electricity consumption shows a right skew, indicating a few high consumption clients.")
print("2. There are missing values in some columns; further cleaning may be required.")
print("3. The correlation heatmap indicates that higher electricity consumption is associated with higher churn rates.")
print("4. Churn rates vary by sales channel, suggesting targeted strategies may be beneficial.")
```

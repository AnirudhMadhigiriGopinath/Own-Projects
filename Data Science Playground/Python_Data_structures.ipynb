```python

#Dictionaries
def dictionary_functions():
        countries = ['spain', 'france', 'germany', 'norway']
        capitals = ['madrid', 'paris', 'berlin', 'oslo']

        europe = {countries[i] : capitals[i] for i in range(len(countries))}
        europe['India'] = 'new delhi'
        for country, capital in europe.items():
                print("The capital of {country} is {capital}".format(country=country, capital=capital))
    
dictionary_functions()

#Lists
def list_functions():
        
        countries = ['spain', 'france', 'germany', 'norway']
        capitals = ['madrid', 'paris', 'berlin', 'oslo']
        countries.append('india')
        capitals.append('new delhi')
        countries.pop(3)
        capitals.pop(3)
        for country, capital in zip(countries, capitals):
                print(f"The capital of {country} is {capital}")
        print(f"Length of countries list: {len(countries)}")
        print(countries[2], capitals[2])
        countries.insert(2, "Netherlands")
        print(countries[0:3])
        cnc = countries + capitals
        print(cnc)
        print(countries.count('france'))
        if 'india' in countries:
                print("capital is New Delhi")
list_functions()

#sets

def set_operations():
    countries = {'spain', 'france', 'germany', 'norway'}
    capitals = {'madrid', 'paris', 'berlin', 'oslo'}
    
    print("Initial Sets:")
    print(f"Countries: {countries}")
    print(f"Capitals: {capitals}")
    print()
    countries.add('italy')
    capitals.add('rome')
    print("After Adding 'italy' and 'rome':")
    print(f"Countries: {countries}")
    print(f"Capitals: {capitals}")
    print()

    countries.remove('france') 
    capitals.discard('paris')  
    print("After Removing 'france' and Discarding 'paris':")
    print(f"Countries: {countries}")
    print(f"Capitals: {capitals}")

    union_set = countries.union(capitals)
    print("Union of Countries and Capitals:")
    print(union_set)

    intersection_set = countries.intersection(capitals)
    print("Intersection of Countries and Capitals (Common Elements):")
    print(intersection_set)


    difference_set = countries.difference(capitals)
    print("Difference (Countries - Capitals):")
    print(difference_set)


    sym_diff_set = countries.symmetric_difference(capitals)
    print("Symmetric Difference (Countries ^ Capitals):")
    print(sym_diff_set)

    print(f"Is 'spain' in countries? {'spain' in countries}")
    print(f"Is 'paris' in capitals? {'paris' in capitals}")

    print("Iterating through countries:")
    for country in countries:
        print(country)

    print(f"Number of countries: {len(countries)}")
    print(f"Number of capitals: {len(capitals)}")


    frozen_set_example = frozenset(countries)
    print("Frozen Set Example:")
    print(frozen_set_example)
set_operations()

        
```
